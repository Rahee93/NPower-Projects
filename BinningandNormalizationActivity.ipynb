```python
import pandas as pd
import matplotlib.pyplot as plt
# Load data from an Excel file
file_path = 'auto-mpg .xlsx'  # Replace with your file path
df = pd.read_excel(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
#1. Drop column Origin from the dataset.
# Drop the 'Origin' column
df_dropped = df.drop(columns='origin')

# Display the updated DataFrame
print("\nDataFrame after dropping the 'origin' column:")
print(df_dropped)
```

```python
#2. Find out all the missing values in the dataset and replace it with its most appropriate replacement.

# Check for missing values in each column
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# Replace missing values based on the type of data
for column in df.columns:
    if df[column].isnull().sum() > 0:  # Check if the column has missing values
        if df[column].dtype in ['float64', 'int64']:
            # Numerical columns: replace with mean or median
            df[column].fillna(df[column].median(), inplace=True)  
        elif df[column].dtype == 'object':
            # Categorical columns: replace with mode
            df[column].fillna(df[column].mode()[0], inplace=True) 

# Check if there are any missing values left
missing_values_after = df.isnull().sum()
print("\nMissing values after replacement:")
print(missing_values_after)
```

```python
#3. Find and remove duplicate entries for the column ‘ CAR Number’

# Find and remove duplicate entries based on the 'CAR Number' column
df_unique = df.drop_duplicates(subset='CAR Number')

# Display the DataFrame after removing duplicates
print("\nDataFrame after removing duplicates:")
print(df_unique)
```

```python
#5. In our dataset, the fuel consumption column is "mpg" and is represented by mpg
#(miles per gallon) unit. Assume we are developing an application in a country
#that accepts fuel consumption with the L/100km standard. change the name of
#the column to "L/100km"

# Rename the 'mpg' column to 'L/100km'
df.rename(columns={'mpg': 'L/100km'}, inplace=True)

# Display the updated DataFrame
print("\nUpdated DataFrame with the column name:")
print(df)
```

```python
#6. Normalize the column “Weight” so that the values range from 0 to 1
# Normalize the 'Weight' column using the Min-Max Normalization Method
df['weight'] = (df['weight'] - df['weight'].min()) / (df['weight'].max() - df['weight'].min())

# Display the updated DataFrame
print("\nUpdated DataFrame with normalized 'weight' column:")
print(df)
```

```python
#7. Normalization is the process of transforming values of several variables into a
#similar range. Typical normalizations include scaling the variable so the variable
#average is 0, scaling the variable so the variance is 1, or scaling variable so the
#variable values range from 0 to 1

# using Z-Score Normalization
df['Weight_Z-score'] = (df['weight'] - df['weight'].mean()) / df['weight'].std()
# Display the DataFrame with all normalized columns
print("DataFrame with Normalized Columns:")
print(df)
```

```python
#8. Approach: replace original value by (original value)/(maximum value)

# Normalize the 'Weight' column by dividing each value by the maximum value
max_weight = df['weight'].max()
df['Weight_Normalized'] = df['weight'] / max_weight

# Display the updated DataFrame
print("\nUpdated DataFrame with normalized 'Weight' column:")
print(df)
```

```python
#9. In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, it
#has 57 unique values. What if we only care about the price difference between
#cars with high horsepower, medium horsepower, and little horsepower (3
#types)? Can we rearrange them into three ‘bins' to simplify analysis?

# Define the bins and labels
bins = [0, 100, 200, 300] 
labels = ['Low Horsepower', 'Medium Horsepower', 'High Horsepower']

# Display the updated DataFrame
print("Updated DataFrame with Horsepower Categories:")
print(df)
```

```python
#10. We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins

# Use the cut method to segment the 'horsepower' column into 3 bins
df['Horsepower_Category'] = pd.cut(df['horsepower'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame with the new categorical column
print("Updated DataFrame with Horsepower Categories:")
print(df)
```

```python
# 11. Use a histogram to visualize the distribution of bins we created above.
# Create a histogram to visualize the distribution of horsepower categories
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(df['horsepower'], bins=4, color='skyblue', edgecolor='black')


# Adding titles and labels
plt.title('Distribution of Horsepower Categories')
plt.xlabel('Horsepower Category')
plt.ylabel('Number of Cars')
plt.xticks(rotation=1) 
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
from scipy.stats import zscore
#  12. Detect outliers using Z-score and remove them
# Drop the Z-Score column as it is no longer needed

# Calculate Z-scores for the 'Horsepower' column
df['Z-Score'] = zscore(df['horsepower'])

# Define a threshold for identifying outliers
threshold = 3

# Identify outliers based on Z-score
outliers = df[(df['Z-Score'] > threshold) | (df['Z-Score'] < -threshold)]

# Print outliers (optional)
print("Outliers:")
print(outliers)

# Remove outliers from the DataFrame
df_cleaned = df[(df['Z-Score'] <= threshold) & (df['Z-Score'] >= -threshold)]

# Drop the Z-Score column as it is no longer needed
df_cleaned = df_cleaned.drop(columns=['Z-Score'])

# Display the cleaned DataFrame
print("\nCleaned DataFrame (without outliers):")
print(df_cleaned)
```
